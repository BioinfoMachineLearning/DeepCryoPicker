program to quickly batch rename files to desired format.

set path to directory, directory must have folders titled to dataset. THERE MUST NOT BE ANYTHNIG BESIDES FOLDERS in this directory

this notebook will gointo each of that directories sub directories and rename all files in it to the desired naming convention.

change new_file_name to desired naming conevention.
currently set up to renmae to "'folder_name'_00X" where x is its file number within the subdirectory.

```python
import os
#drive path to directory containg datasets
path = "..\Data Sets\Raw Datasets"

# Loop through all folders in the directory
for folder_name in os.listdir(path):
    folder_path = os.path.join(path, folder_name)
    if os.path.isdir(folder_path):
        # Loop through all files in the folder
        for i, file_name in enumerate(os.listdir(folder_path)):
            file_path = os.path.join(folder_path, file_name)
            if os.path.isfile(file_path):
                # Construct the new file name
                new_file_name = f"{folder_name}_{i+1:03}"
                new_file_path = os.path.join(folder_path, new_file_name)
                # Rename the file
                os.rename(file_path, new_file_path)
```
