**MRCtoPNG**
!! this is depricated as the mrc to png conversion has now been implmented into the preprocessign pipeline (basically running preprocessing will convert .mrc to .png autamatically if it is needed)

converts from MRC file format to PNG. The datasets for this project are in MRC file format, but we need them to be in PNG.

This utiliy will copy all contents of Source directory to Destination directoy.

Once it has been converted it will move the Destination directory to the specified location in project folder. 

To run, set up a Target Directory, this directory must follow this structure:


> Source Directory

>>Dataset Folder

>>Dataset Folder

Dataset folder should be titled as the name of the data set

For Destination Directory, you just need to create an empty directory

Update the vlaues for input_folder to Source Directory, and output_folder to Destination Directory in cell 2.

optionally, you can run cell 4 to copy the converted Destination Directory to where ever you want, just update path to desired location

```python
import os
import mrcfile
import imageio
import shutil
```

```python
# Path to the Source Directory containing MRC files
input_folder = '/content/mrc-folder'

# Path to the Destination Directory for PNG files
output_folder = '/content/png-output'
```

```python
# Loop through all files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.mrc'):
        # Construct full input and output file paths
        input_file = os.path.join(input_folder, filename)
        output_file = os.path.join(output_folder, os.path.splitext(filename)[0] + '.png')

        # Open the MRC file and read the data
        with mrcfile.open(input_file) as mrc:
            data = mrc.data

        # Save the data as a PNG image
        imageio.imwrite(output_file, data)
```
